## Ejercicio 1

Crea una función que busque todos los divisores del número entero positivo dado por parámetro y devuelva
una lista con todos los divisores de dicho número.

```python
print("Este algoritmo te entregará todos los divisores del número que suministres.")

def divisors(num):

  div = []

  for i in range(num):
    if num % (i + 1) == 0:
      div.append(i + 1)

  return div

num = int(input("Ingresa el número para entregarte sus respectivos divisores: "))

print("La lista de divisores del número {} suministrado es {}.".format(num, divisors(num)))
```

## Ejercicio 5

Crea una función que devuelva el MCD (máximo común divisor) de 2 números proporcionados por parámetro.

**PISTA:** Aprovecha la función que calcula el mayor entre dos números dados. También necesitarás una función
que calcula el menor entre dos números dados.

```python
print("Este algoritmo te entregará el MCD de dos números que suministres.")

def may(a, b):
  if a > b:
    return a
  elif a < b:
    return b
  else:
    return a

def min(a, b):
  if a > b:
    return b
  elif a < b:
    return a
  else:
    return a

a = int(input("Ingresa el primer número que se va a comparar: "))
b = int(input("Ingresa el segundo número que se va a comparar: "))

def MCD(a, b):

  r = 0
  bigger = may(a, b)
  lower = min(a, b)

  while lower > 0:
    r = lower
    lower = bigger % lower
    bigger = r
  return bigger

print("El MCM entre {} y {} es {}.".format(a, b, MCD(a, b)))
```

## Ejercicio 6

Crea una función que devuelva el MCM (mínimo común múltiplo) de 2 números proporcionados por
parámetro.

**PISTA:** Aprovecha la función que calcula el MCD de dos números del ejercicio anterior y la función que
calcula el valor absoluto de un número. Recuerda la formula,

$$ mcm(a, b) = \frac{\begin{vmatrix}
ab
\end{vmatrix}}{MCD(a, b)}$$

```python
print("Este algoritmo te devolverá el mcm de dos números que ingreses.")


def may(a, b):
  if a > b:
    return a
  elif a < b:
    return b
  else:
    return a

def min(a, b):
  if a > b:
    return b
  elif a < b:
    return a
  else:
    return a

a = int(input("Ingresa el primer número que se va a comparar: "))
b = int(input("Ingresa el segundo número que se va a comparar: "))

def MCD(a, b):

  r = 0
  bigger = may(a, b)
  lower = min(a, b)

  while lower > 0:
    r = lower
    lower = bigger % lower
    bigger = r
  return bigger

def absolute_value(x):

  if x >= 0:
    return x
  else:
    return -x

def mcm(a, b):
  return absolute_value(a * b) // MCD(a, b)

print("El mcm de los números {} y {} es {}.".format(a, b, mcm(a, b)))
```

## Ejercicio 7

Crea una función que dada una palabra devuelva si es palíndroma.

```python
print("Este algoritmo te informará si la palabra suministrada es o no palíndroma.")

s = input("Ingresa la palabra correspondiente: ")

def palindroma(caract):
  word = ""
  caract = caract.lower()
  for i in range(len(caract)):
    word += caract[len(caract) - i - 1]

  if caract == word:
    print("La palabra {}, efectivamente, es palíndroma.".format(caract))
    return word

print(palindroma(s))
```

## Ejercicio 9

Crea una función que dada una lista de palabras por parámetro, devuelva un diccionario que contenga
cuántas son de longitud par y cuántas de longitud impar.

```python
print("Este algoritmo te indicará cuántas palabras que ingreses son de longitud par y de longitud impar.")

def cont(*words):
  even = 0
  noteven = 0

  for i in words:
    if len(i) % 2 == 0:
      even += 1
    else:
      noteven += 1

  word = {"Pares" : even, "Impares" : noteven}

  return word


value = cont("Azucar", "Agua", "Arequipe", "Arrecife", "Arnoldo", "Gatitos")

print("El conjunto de palabras con longitud par e impar es", value)
```

## Ejercicio 10

Crea una función que dado un string por parámetro cuente cuántas veces sale cada caracter en dicho string
y devuelva toda esa información en un diccionario.

```python
print("Este algoritmo te indicará cuántas veces aparece un caracter en el string que suministres.")

def cont(s):
  count = {}
  s = s.lower()

  for c in s:
    conta = 0
    for j in s:
      if c == j:
        conta += 1
    count.setdefault(c, conta)

  return count

s = input("Inserte una frase: ")

print("El diccionario con la cantidad de veces que aparece cada caracter es.", cont(s))
```

## Ejercicio 11

Crea una función que calcule el promedio de $n$ números suministrados por el usuario. Recuerde que la formula viene dada así:

$$ \bar{x} = \frac{\sum_{i = 1}^{n}x_{i}}{n} $$

```python
print("Este algoritmo te devolverá el promedio de los números que suministres.")

def average(*n):
  sum = 0
  for i in n:
    sum += i
  aver = sum / len(n)
  return aver

value = average(4, 4, 5, 4)
print("El promedio de los números suministrados es", value)
```

## Ejercicio 12

Crea una función que con el Teorema de Pitágoras y los dos catetos suministrados por el usuario, se calcule la hipotenusa. Recuerda que la fórmula se define así,

$$ h^{2} = c_{1}^{2} + c_{2}^{2} $$

Simplificando,

$$ h = \sqrt{c_{1}^{2} + c_{2}^{2}} $$

```python
from numpy import sqrt

print("Este algoritmo calculará la hipotenusa de un triángulo rectángulo con los catetos que suministres.")

def Pythagoras_Theorem(c1, c2):
  h = sqrt(c1 ** 2 + c2 ** 2)
  return h

c1 = float(input("Ingresa el cateto 1 del triángulo: "))
c2 = float(input("Ingresa el cateto 2 del triángulo: "))

print("El valor de la hipotenusa del triángulo es {}.".format(Pythagoras_Theorem(c1, c2)))
```

## Ejercicio 13

Crea una función que calcule la distancia entre dos puntos en el plano xy. Recuerda que la formula se define así,

$$ d(x, y) = \sqrt{(x_{1}-x_{2})^{2} + (y_{1}-y_{2})^{2}} $$

```python
from numpy import sqrt

print("Este algoritmo te entregará la distancia entre dos puntos que suministres.")

x1 = float(input("Ingresa la coordenada x1 del punto 1: "))
y1 = float(input("Ingresa la coordenada y1 del punto 1: "))
P1 = (x1, y1)
print("El punto 1 es", P1)

x2 = float(input("Ingresa la coordenada x2 del punto 2: "))
y2 = float(input("Ingresa la coordenada y2 del punto 2: "))
P2 = (x2, y2)
print("El punto 2 es", P2)

def distance_function(P1, P2):
  d = sqrt((P1[0] - P2[0])**2 + (P1[1] - P2[1])**2)
  return d

print("La distancia entre el punto {} y el punto {} es {}.".format(P1, P2, distance_function(P1, P2)))
```

## Ejercicio 14

Crea un programa que gráfique una función polinómica. Recuerda que una función de una variable se define de la forma,

$$ y = f(x) $$

Siendo la función,

$$ f(x) = x^{3} + 5x^{2} - 3x + 2 $$

Utiliza el módulo matplotlib para gráficar la función polinómica determinada.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**3 + 5 * x**2 - 3 * x + 2

#Se utiliza el método linspace de numpy
#linspace genera 1000 puntos entre -10 y 10 para su respectiva gráfica.
x = np.linspace(-10, 10, 1000)
y = [f(x) for x in x]

plt.grid()
plt.axvline(x=0, color='black', linestyle='-') #Eje y
plt.axhline(y=0, color='black') #Eje x
plt.xlim(-10,10)
plt.ylim(-500,1500)
plt.title("f(x) = x**3 + 5 * x**2 - 3 * x + 2")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, color='blue')
```

## Ejercicio 15

Crea un programa que gráfique una función a trozos. Se te suministra la siguiente función a trozos:

$$ f(x) = \left\{\begin{matrix}
x^{2} \: \: \textrm{si} \: \: x >  0\\
4 - x \: \: \textrm{si} \: \: x \leq   0
\end{matrix}\right. $$


Utiliza el módulo matplotlib para gráficar la función a trozos suministrada.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  if x > 0:
    return x**2
  else:
    return 4 - x

#Se utiliza el método linspace de numpy
#linspace genera 1000 puntos entre -10 y 10 para su respectiva gráfica.
x = np.linspace(-10, 10, 1000)
y = [f(x) for x in x]

plt.grid()
plt.axvline(x=0, color='black', linestyle='-') #Eje y
plt.axhline(y=0, color='black') #Eje x
plt.title("Función a trozos")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, color='blue')
```
