## Ejercicio 4

Dado el array 2D [["ala", "delfín", "arroz"], ["cena", "kayak", "picnic"], ["hoja", "gato",
"elle"]], busca las palabras que sean palíndromos.
PISTA: Necesitarás una función universal.

```python
a = np.array([["ala", "delfín", "arroz"], ["cena", "kayak", "picnic"], ["hoja", "gato", "elle"]])

def IsPalindrome(s):
  s = s.lower()
  word1 = ""
  word2 = ""
  for i in s:
    word1 += i
  for i in s[::-1]:
    word2 += i
  if word1 == word2:
      return True
  return False

my_palindrome = np.frompyfunc(IsPalindrome, 1, 1)
print(my_palindrome(a))
```

## Ejercicio 7

Crea un minijuego que piense un array 1D aleatorio de 5 elementos. Los elementos serán números enteros
entre 1 y 20. Para cada intento, primero tendrá que pedir 5 números enteros del 1 al 20. El juego nos tendrá
que decir cuántos números del array hemos acertado y cuántos números del array están bien colocados. El
juego no termina hasta que no se hayan acertado todos los números pensados en la posición correspondiente.
Al acabar, se tendrá que mostrar la frase: “¡¡¡ENHORABUENA!!! Has acertado el array que había pensado:
{}”


```python
a = np.array([random.randint(1, 21), random.randint(1, 21), random.randint(1, 21), random.randint(1, 21), random.randint(1, 21)])
b = [0] * len(a)

while not np.array_equal(b, a):
  b = []
  for i in range(5):
    b.append(int(input("Piense un número entre 1 y 20: ")))
  b = np.array(b)

  for i in range(5):
    if b[i] == a[i]:
      print("Has acertado el número", a[i] , "en la posición", i)

  for i in range(5):
    if b[i] in a:
      print("Has acertado el número", b[i])
    else:
      print("No has acertado el número", b[i])
else:
  print("¡¡¡ENHORABUENA!!! Has acertado el array que había pensado:", a)
```

## Ejercicio 8

A partir de la lista [-3, -2, -1, 0, 1, 2, 3], genera dos arrays aleatorios 2D de size = (3, 4) y
súmalos.


```python
import numpy as np

# Lista original de números con un elemento adicional para tener 12 elementos
lista = np.array([-3, -2, -1, 0, 1, 2, 3, 0, 0, 0, 0, 0])

# Generamos dos arreglos 2D aleatorios a partir de la lista original
arreglo1 = np.random.permutation(lista).reshape(3, 4)
arreglo2 = np.random.permutation(lista).reshape(3, 4)

# Sumamos los dos arreglos
resultado = arreglo1 + arreglo2

print("Arreglo 1:\n", arreglo1)
print("Arreglo 2:\n", arreglo2)
print("Suma de Arreglo 1 y Arreglo 2:\n", resultado)
```

## Ejercicio 9

Crea una función universal que dado un array de strings devuelva el número total de vocales de cada string.


```python
import numpy as np

# Función para contar las vocales en una cadena
def count_vowels(s):
    count = 0
    vowels = set("aeiouAEIOU")
    for char in s:
        if char in vowels:
            count += 1
    return count

# Crear una función vectorizada a partir de count_vowels
count_vowels_vec = np.vectorize(count_vowels)

# Pedimos al usuario que ingrese el número de palabras
n = int(input("¿Cuántas palabras quieres introducir?: "))

# Lista para almacenar las palabras
list1 = []
for i in range(n):
    a = input("Introduce una palabra: ")
    list1.append(a)

# Convertimos la lista en un array de numpy
AB = np.array(list1)

# Aplicamos la función vectorizada al array
my_vowels = count_vowels_vec(AB)

# Imprimimos el resultado
print("Número de vocales en cada palabra:", my_vowels)
```

## Ejercicio 10

Crea una función universal que dados un array 1D de valores string y conjunto de strings, compruebe para
cada elemento del array 1D, si está presente o no en el conjunto.

```python
import numpy as np

# Función para comprobar si un string está en el conjunto
def is_in_set(s, string_set):
    return s in string_set

# Crear una función vectorizada a partir de is_in_set
is_in_set_vec = np.vectorize(is_in_set, excluded=['string_set'])

# Definir el conjunto de strings
string_set = {"apple", "banana", "cherry"}

# Pedimos al usuario que ingrese el número de palabras
n = int(input("¿Cuántas palabras quieres introducir?: "))

# Lista para almacenar las palabras
list1 = []
for i in range(n):
    a = input("Introduce una palabra: ")
    list1.append(a)

# Convertimos la lista en un array de numpy
AB = np.array(list1)

# Aplicamos la función vectorizada al array
present_in_set = is_in_set_vec(AB, string_set=string_set)

# Imprimimos el resultado
print("Comprobación de presencia en el conjunto:", present_in_set)
```
