## Ejercicio 1

Crea un programa que pida un número entero positivo por teclado y que cree un diccionario cuyas claves
sean desde el número 1 hasta el número indicado. Los valores de cada clave serán las propias claves elevadas
al cubo.

```python
print("Este algoritmo creará un diccionario con cada número del 1 hasta el número que introduciste como clave y como valor se tendrá \nese mismo número al cubo.")
print("Debes ingresar un número entero positivo.")
n = int(input("Ingresa el número: "))
dicc = {}
cont = 0
dicti = []
clave = []
value = []

if n > 0:
  print("Sí seleccionaste correctamente el número.")
  for i in range(n):
    dicti.append([])
    cont += 1
    clave.append(cont)
    num = cont ** 3
    value.append(num)
    dicti[i].append(clave[i])
    dicti[i].append(value[i])

  dicti = dict(dicti)
  print("Este es el diccionario con todos los valores", dicti)

else:
  print("No seleccionaste un número entero positivo como se te indicó.")
```

## Ejercicio 9

Crea un programa que pida el número de palabras a introducir. Crear un diccionario de clave la palabra y
de valor el número de vocales de la palabra.

```python
print("Este algoritmo te entregará un diccionario con cada palabra como clave y su respectiva longitud como valor.")

n = int(input("Ingresa el número de palabras que vas a introducir: "))

dicc = {}

for i in range(n):
  cont = 0
  word = str(input("Ingresa la palabra {}: ".format(i + 1)))
  word_proof = word.lower()
  for c in word_proof:
    if c == "a" or c == "e" or c == "i" or c == "o" or c == "u":
      cont += 1
  dicc.setdefault(word, cont)

print("La lista ordenada de palabras con su respectivo número de vocales es: \n", dicc)
```

## Ejercicio 16

Escribir un programa que cree un diccionario de traducción español-inglés. El usuario introducirá las palabras en español e inglés separadas por dos puntos, y cada par <palabra>:<traducción> separados por comas. El programa debe crear un diccionario con las palabras y sus traducciones. Después pedirá una frase en español y utilizará el diccionario para traducirla palabra a palabra. Si una palabra no está en el diccionario debe dejarla sin traducir.

```python
print("Este algoritmo te mostrará un diccionario del español al inglés de palabras que tú conozcas.")

dicc = {}


answer = str(input("¿Deseas ingresar datos a un diccionario? "))

while answer != "no":
  key = str(input("Indica la palabra en español: "))
  value = str(input("Indica su traducción en inglés: "))
  dicc.setdefault(key, value)
  answer = str(input("¿Deseas continuar ingresando datos al diccionario? "))
  answer = answer.lower()

s = str(input("Ingresa la frase en español: "))

for i in s.split():
  if i in dicc:
    s = s.replace(i, dicc[i])

print("La frase traducida al inglés es:", s)
```

## Ejercicio 20

Realice un programa que cree un diccionario con todos los valores de la Distribución de Probabilidad Normal Estándar; esto es, los valores de z como clave y las probabilidades como valores. Recuerde que la Distribución de Probabilidad Normal es probablemente la distribución continua más importante tanto en estadística teórica como aplicada. Esta es una distribución que refleja muy bien como la mayoría de datos se concentran sobre la media de la población. Su forma es de campana (conocida como Campana de Gauss) y es simétrica. La ecuación de la Distribución de Probabilidad Normal es la siguiente:


$$ f(x)= \frac{1}{\sigma \sqrt{2\pi }} e^{\frac{-1}{2} \left ( \frac{x-\mu}{\sigma } \right )^{2}} $$

Donde,

$ \mu = \mathrm{Media} $

$ \sigma  = \mathrm{Desviacion\: Estandar} $

$ x = \mathrm{Observacion} $

Ahora bien, para representar la tabla normal estándar se debe resolver la siguiente integral transformada:


$$ \phi (z) = \frac{1}{\sqrt{2\pi }} \int_{-\infty}^{z} e^{\frac{-t^{2}}{2}}\:dt $$


Para realizar la integral con métodos númericos utilice el módulo scipy.

```python
import numpy as np
import scipy as sp
from scipy.integrate import quad

print("Este algoritmo te entregará la tabla de distribución normal de forma que P(z <= Z) de la curva de Distribución Normal.")
#Los valores de la tabla de la distribución normal comienzan desde -3.99
x = -3.99
normal_table = {}

def f(t):
  return (1 / np.sqrt(2 * np.pi)) * np.exp(-t**2 / 2)

def Phi(z):
  prob = quad(f, -np.inf, z)[0]
  return prob

while x < 4:
  value = Phi(x)
  normal_table.setdefault(x, value)
  x += 0.01

print("La tabla de distribución normal para P(z <= Z) es:")
print(normal_table)
```
