## Ejercicio 10

Dada una ecuación de segundo grado $ax^{2} + bx + c = 0$, sabemos que las soluciones vienen dadas por:

$$ x = \frac{-b\pm\sqrt{b^{2}-4ac}}{2a} $$

Siempre y cuando $a \neq 0.$
El número de soluciones dentro del conjunto de los números reales $ \mathbb{R} $ viene dado en función del signo que tome el discriminante, $ \bigtriangleup = b^{2} - 4ac.$ Si $ \bigtriangleup > 0$, entonces tendremos dos soluciones diferentes en  $ \mathbb{R}, $ $ \left (x = \frac{-b+\sqrt{\bigtriangleup}}{2a} \hspace{0.2cm} y \hspace{0.2cm}x = \frac{-b-\sqrt{\bigtriangleup}}{2a} \right) $. Si $ \bigtriangleup = 0 $, entonces tendremos dos soluciones que serán iguales $ \left ( x = \frac{-b}{2a}  \right )$. Si $ \bigtriangleup < 0 $, entonces no existe solución dentro del conjunto de los números reales para esta ecuación de segundo grado.

Con toda esta información, haz que el usuario introduzca los valores de los coeficientes $ a, b, c $. Con ello,

1. Comprueba que el coeficiente $a$ es distinto de 0.
2. En función del discriminante, calcula cuántas soluciones va a tener la ecuación de segundo grado $ax^{2} + bx + c = 0$.
3. Devuelve por pantalla el resultado en cada caso (tanto el número de soluciones en los números reales como el valor de éstas).

PISTA: Para calcular la raíz cuadrada, vas a necesitar la función $math.sqrt()$ de la librería $math$.

```python
from math import sqrt

print("Este es un algoritmo matemático que te dará las soluciones para cualquier ecuación cuadrática. Para que este funcione, debes insertar los coeficientes a, b y c.")
a = float(input("Inserta el coeficiente a: "))
b = float(input("Inserta el coeficiente b: "))
c = float(input("Inserta el coeficiente c: "))

if a != 0:
  D = b**2 - 4 * a * c
  if D > 0:
    print("El polinomio tiene dos soluciones en los números reales.")
    sq = sqrt(b**2 - 4 * a * c)
    s1 = (-b + sq) / (2*a)
    s2 = (-b - sq) / (2*a)
    print("Las solución x1 es {}. La solución x2 es {}.".format(s1, s2))
  elif D == 0:
    print("El polinomio tiene una única solución que cuenta por dos.")
    s = -b / (2*a)
    print("La solución x es {}.".format(s))
  else:
    print("El polinomio de grado dos no tiene soluciones en los números reales.")
else:
  print("La ecuación que insertaste no es una ecuación cuadrática.")
```

## Ejercicio 20

A un trabajador le descuentan de su sueldo el 10% si su sueldo es menor o igual a 1000. Por encima de 1000 y hasta 2000 el 5% del adicional, y por encima de 2000 el 3% del adicional. Calcular el descuento y sueldo neto que recibe el trabajador dado su sueldo.

*NOTA:* Hacer este ejercicio preferiblemente con una función a trozos.

```python
# Se define la función a trozos por medio de los condicionales. Más adelante se verán métodos más complejos para trabajar con funciones matemáticas.
# Más adelante se ahondará en el concepto de función matemática y su gráfica.

def f(x):
  if x <= 1000:
    return 0.9 * x
  elif x > 1000 and x <= 2000:
    return (x - 1000) * 0.95 + 0.9 * 1000
  elif x > 2000:
    return (x - 2000) * 0.97 + 0.9 * 1000 + 0.95 * 1000

def g(x):
  if x <= 1000:
    return 0.1 * x
  elif x > 1000 and x <= 2000:
    return (x - 1000) * 0.05 + 0.1 * 1000
  elif x > 2000:
    return (x - 2000) * 0.03 + 0.1 * 1000 + 0.05 * 1000

print("Por medio de este algoritmo se sabrá cuánto se le descuenta y se le paga al empledo en función de su sueldo. Se cobran parafiscales.")
salary = float(input("Ingresa tu salario: "))
pay = f(salary)
discount = g(salary)
print("Tu pago será de {} y tu descuento es de {}.".format(pay, discount))
```

*   Si el valor de la compra esta entre $\$100.000$ y $\$200.000$ el descuento que aplica es del 5%.
*   Si el valor de compra es mayor que $\$200.000$ el descuento que aplica es del 7%

## Ejercicio 37

Realice un algoritmo que pruebe si un año es bisiesto o no.
*   Un año es bisiesto si es divisible entre cuatro pero no es múltiplo de cien a no ser que lo sea de 400.



```python
year = int(input("Introduce el año a probar si es o no bisiesto: "))

if year % 4 == 0:
  if year % 100 == 0:
    if year % 400 == 0:
      print("El año {} es bisiesto.".format(year))
    else:
      print("El año {} no es bisiesto.".format(year))
  else:
    print("El año {} es bisiesto.".format(year))
else:
  print("El año {} no es bisiesto.".format(year))
```
