## Ejercicio 8

Realice un programa que lea a 10 usuarios, de manera que a cada uno le pida el valor de su salario. Al final usted debe indicar cuál es el promedio de los salarios.

```python
cont = 1
sum = 0

while cont <= 10:
  salario = float(input("Ingrese su salario: "))
  sum += salario
  cont += 1
else:
  promedio = sum / 10
  print(f"El promedio de los salarios es {promedio}")
```

## Ejercicio 9

Una frutería recibe N proveedores con X bultos de naranja, cada bulto de naranja es comprado a 25 mil pesos si son menos de 10, o a 28 mil en caso contrario. Realice un programa que calcule para un día, cuántos bultos de naranja se compraron, cuanto costaron todos los bultos de naranja y el promedio del costo de los bultos. Pida los valores de N y X al usuario.

```python
N = int(input("¿A cuántos proveedores se les compra bultos de naranja? "))
cont = 0
bultos = 0
costo = 0

while cont < N:
  X = int(input(f"¿Cuántos bultos de naranja se comprarán al proveedor {cont + 1}? "))
  bultos += X
  if X < 10:
    costo += 25000 * X
  else:
    costo += 28000 * X
  cont += 1
else:
  promedio = costo / N
  print(f"Se le compro a {N} proveedores")
  print(f"Se le gasto ${costo} en bultos de fruta")
  print(f"El promedio de lo gastado es ${promedio}")
```

## Ejercicio 10

Realice un programa que genere un número aleatorio entre 0 y 100 (el número aleatorio se generará una sola vez). Luego el usuario digitará un valor hasta que adivine el número que se generó aleatoriamente. Cada que el usuario se equivoque, el programa le deberá decir si el número a adivinar es mayor, o menor que aquel que fue digitado por el usuario. Al acertar el número, en pantalla debe aparecer un mensaje que diga que acertó y cuántos intentos necesitó “acertaste el número y necesitaste # intentos”.

```python
from random import randint

ran_num = randint(0, 100)
num = int(input("Adivina un número entre 0 y 100: "))
n = 1

while num != ran_num:
  print("No acertaste con el número")
  if num > ran_num:
    print("El número que indicaste es mayor que el número generado aleatoriamente")
  else:
    print("El número que indicaste es menor que el número generado aleatoriamente")

  num = int(input("Vuelve a adivinar el número entre 0 y 100: "))
  n += 1
else:
  print(f"¡Felicidades! Has acertado el número que se generó. Lo lograste en {n} intentos.")
```

## Ejercicio 11

Realice un programa que le pida al usuario 10 números enteros (debe leerlos uno a uno). Adicionalmente, solamente la primera vez que el usuario escriba un cinco usted debe felicitarlo. Si el usuario nunca escribe cinco, entonces usted nunca lo felicitaría.

```python
cont = 1
cont5 = 0
condition_five = True

while cont <= 10:
  num = int(input(f"Ingresa el número {cont} de los 10 solicitados: "))
  if num == 5:
    cont5 += 1
  if cont5 == 1 and condition_five == True:
    print('¡Felicitaciones!')
    condition_five = False
  cont += 1
```

## Ejercicio 12

Pida al usuario dos números enteros, y realizando un ciclo, vaya acumulando el valor multiplicando la respuesta por el segundo número, por ejemplo, si el primer número es 2 y el segundo es 3, los valores serían (6, 18, 54….). El ciclo se debe detener cuando llegue a un número que supere el valor de 10.000. Muestre cuál fue el número con que terminó la serie.

```python
print("Se te pedirán dos números. Estos se multiplicarán hasta superar el número 10.000. Se te indicará cuál fue el número que terminó la serie")
num = int(input(f"Ingresa un número: "))
num2 = int(input(f"Ingresa otro número: "))

while num <= 10000:
    num = num * num2
    print(f"El siguiente número de la serie es {num}")
else:
  print(f"El número {num} terminó la serie.")
```

## Ejercicio 13

Realice un programa que le pida al usuario 10 números enteros positivos (debe leerlos uno a uno), y luego retorne el menor de los 10 números digitados por el usuario.

```python
min = None
cont = 1
print("Se te mostrará el número menor de los 10 números que ingreses.")
while cont <= 10:
  num = int(input("Ingresa un número: "))
  if min is None or num < min:
    min = num
  cont += 1
else:
  print(f"El número menor es {min}")
```

## Ejercicio 14

Realice un programa que pida al usuario números enteros (debe leerlos uno a uno) hasta que el usuario digite menos uno (-1), luego muestre en pantalla tanto el número menor, como el número mayor de los digitados por el usuario (sin tener en cuenta el -1)

```python
condition = False
print("Se te mostrarán los números mayor y menor que ingreses.")
min = None
max = 0
while condition == False:
  num = int(input("Digita un número: "))
  if num == -1:
    break
  if min is None or min > num:
    min = num
  if max < num:
    max = num

print(f"El número mayor fue {max}")
print(f"El número menor fue {min}")
```
