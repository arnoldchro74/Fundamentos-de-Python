# Ciclo For

## Ejercicio 1

Sume los múltiplos del 9 entre el 1 y el 10.000

```python
sum = 0
for i in range(10001):
  if i % 9 == 0:
    sum += i
print(sum)
```

## Ejercicio 2

Multiplique los números que sean múltiplos de 3 y 5 a vez entre el 1 y el 100

```python
mul = 1

for i in range(1, 101):
  if i % 3 == 0 and i % 5 == 0:
    mul *= i

print(f"La multiplicación entre los números múltiplos de 3 y 5 del 1 al 100 es {mul}")
```

## Ejercicio 3

Pregunte al usuario, cuántas veces desea que le pregunte nombres de integrantes de la familia, realice el ciclo que se repita esa cantidad de veces, y muestre en pantalla cuántas veces escribió “Camila”.

```python
n = int(input("¿Cuántos integrantes de la familia quiere ingresar?: "))
cont = 0

for i in range(n):
  name = input("Ingrese el nombre del integrante de la familia: ").strip().lower()
  if name == "camila":
    cont += 1
else:
  print(f"Escribiste {cont} veces el nombre de Camila")
```

## Ejercicio 4

Creemos un programa que recibe n cantidad de cadenas de caracteres y que muestra en pantalla sólo aquellas cadenas que tienen como máximo 8 caracteres.

```python
n = int(input("Escribe la cantidad de caracteres que vas a ingresar: "))

for i in range(n):
  word = input("Escribe tu cadena de caracteres: ").strip().lower()
  if len(word) <= 8:
    print(f"La palabra es menor o igual a 8 caracteres. Esta es {word.capitalize()}")
```

## Ejercicio 5

Cree una cadena de caracteres con la frase: “Colombia es un país hermoso con una amplia variedad de sitios turísticos. No te quedes sin conocerlos”.

Cree un programa usando el ciclo for, de manera que diga en pantalla la posición en la que se encuentra la primera vez que aparece la letra “u” y que a la vez muestre en pantalla, cuántas veces se repite la letra “m”

```python
chain = "Colombia es un país hermoso con una amplia variedad de sitios turísticos. No te quedes sin conocerlos"
condition_u = True
cont_m = 0
pos = 0

for i in chain.lower():
  pos += 1
  if i == "u" and condition_u == True:
    print(f"La primer u se encuentra en la posición {pos}")
    condition_u = False
  if i == "m":
    cont_m += 1
else:
  print(f"La letra m se repite {cont_m} veces")
```
