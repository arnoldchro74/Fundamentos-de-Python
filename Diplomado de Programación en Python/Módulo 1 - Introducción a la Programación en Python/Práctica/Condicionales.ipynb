## Ejercicio 1

Dado que una persona para el promedio colombiano se considera alta si mide más de 1.80, realice un programa que le pregunte al usuario su estatura y le diga si es alto o no.

```python
estatura = float(input("Ingrese su estatura: "))
if estatura > 1.80:
  print("Usted es alto")
else:
  print("Usted no es alto")
```

## Ejercicio 2

Realice un programa que le pida al usuario el nombre, de forma tal que si el usuario se llama “Juan” el sistema le diga. Usted es el ganador. De lo contrario, le diga. Usted ha perdido.

```python
nombre = input("Ingrese su nombre: ")
if nombre == "Juan":
  print("Usted es el ganador")
else:
  print("Usted ha perdido")
```

## Ejercicio 3

Solicite 2 variables al usuario, y realice un programa que imprima cuál es la variable menor, cuál es la variable mayor, o en el caso de ser iguales, que lo diga.

```python
A = int(input("Ingresa el valor de A: "))
B = int(input("Ingresa el valor de B: "))
if A == B:
  print("A y B son iguales")
elif A > B:
  print("A es la variable mayor y B es la variable menor")
else:
  print("B es la variable mayor y A es la variable menor")
```
