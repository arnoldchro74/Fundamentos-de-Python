# Tasas de Cambio (USD a COP)

```python
# Importamos las librerías
import pandas as pd
import requests as rq
```

Miramos la página que tiene los datos del dólar más reciente con request,

```python
response = rq.get('https://api.exchangerate-api.com/v4/latest/USD')
print(response.status_code)
```

Si da 200, la página esta funcionando. De lo contrario, daría 404, que no estaría funcionando.

```python
data = response.json()
```

```python
df = pd.DataFrame(list(data["rates"].items()), columns=["Moneda", "Tasa"])
display(df)
```

Ahora miramos los datos de la moneda COP y EUR con respecto al USD

```python
if response.status_code == 200:
    data = response.json()
    date_day = data['date']
    # Obtener las tasas de cambio para COP y EUR desde USD
    usd_to_cop = data['rates']['COP']
    usd_to_eur = data['rates']['EUR']
     # Imprimir las tasas de cambio obtenidas
    print(f"Datos consultados para el día {date_day} ")
    print(f"1 USD = {usd_to_cop} COP")
    print(f"1 USD = {usd_to_eur} EUR")
else:
    print(f"Error al hacer la solicitud: {response.status_code}")
```
