# Utilización de Numpy

```python
# Importar librerias
import numpy as np
```

```python
temperaturas = np.array([[22, 25, 20], [24, 27, 22], [21, 23, 19], [25, 29, 24], [26, 30, 25], [23, 28, 22],
 [24, 26, 21], [22, 25, 20], [27, 31, 26], [28, 32, 27], [29, 33, 28], [21, 24, 19], [20, 23, 18], [19, 22, 17],
  [23, 26, 21], [24, 28, 23], [25, 29, 24], [26, 30, 25], [27, 31, 26], [28, 32, 27], [29, 33, 28], [30, 34, 29],
   [22, 25, 20], [24, 27, 22], [21, 24, 19], [20, 23, 18], [19, 22, 17], [23, 26, 21], [24, 28, 23], [25, 29, 24]])
```

```python
print("Temperaturas Promedio")
print(f"Temperatura ciudad 1: {round(float(np.mean(temperaturas[:][0])), 2)}")
print(f"Temperatura ciudad 1: {round(float(np.mean(temperaturas[:][1])), 2)}")
print(f"Temperatura ciudad 1: {round(float(np.mean(temperaturas[:][2])), 2)}")
```

```python
print("Temperaturas Mínima")
print(f"Temperatura ciudad 1: {round(float(min(temperaturas[:][0])), 2)}")
print(f"Temperatura ciudad 1: {round(float(min(temperaturas[:][1])), 2)}")
print(f"Temperatura ciudad 1: {round(float(min(temperaturas[:][2])), 2)}")
```

```python
print("Temperaturas Máximas")
print(f"Temperatura ciudad 1: {round(float(max(temperaturas[:][0])), 2)}")
print(f"Temperatura ciudad 1: {round(float(max(temperaturas[:][1])), 2)}")
print(f"Temperatura ciudad 1: {round(float(max(temperaturas[:][2])), 2)}")
```

```python
dias_calientes = np.any(temperaturas > 30, axis=1)
print(f"Días con temperaturas mayores a 30 grados: {np.where(dias_calientes)[0] + 1}")
dias_frios = np.all(temperaturas < 20, axis=1)
print(f"Días con temperaturas menores a 20 grados: {np.where(dias_frios)[0] + 1}")
```

```python
temperaturas_f = (temperaturas * 9/5) + 32
print("Temperaturas en °F (primeros 5 días):\n", temperaturas_f[:5])
```

```python
temp_min = np.min(temperaturas)
temp_max = np.max(temperaturas)

temperaturas_norm = (temperaturas - temp_min) / (temp_max - temp_min)
print("Temperaturas normalizadas (0-1):\n", temperaturas_norm[:5])
```
