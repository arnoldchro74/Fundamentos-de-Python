## Ejercicio 1

Realizar un programa que dada una lista de números enteros (puede ser inventada), y muestre en pantalla el valor de la suma de todos sus elementos.

```python
Milista = [3, 6, 9, 12]
sum = 0

for i in Milista:
  sum += i

print(f"La suma de la lista es {sum}")
```

## Ejercicio 2

Realizar un programa que dada una lista de números enteros (puede ser inventada), muestre en pantalla la suma de los elementos que se encuentren en las posiciones pares de la lista.

```python
Milista = [3, 6, 9, 12]
sum = 0

for i in range(len(Milista)):
  if i % 2 == 0:
    sum += Milista[i]

print(f"La suma de la lista con posiciones pares es {sum}")
```

## Ejercicio 4

Realizar un programa, que dada una lista y un número debe mostrar “True” si el valor está en la lista y “False” de lo contrario.

```python
Milista = [3, 6, 9, 12]
numbers = [5, 9]

for i in range(len(numbers)):
  encontrado = False
  for j in range(len(Milista)):
    if numbers[i] == Milista[j]:
      encontrado = True
  print(encontrado)
```

## Ejercicio 5

Basado en el punto anterior, ahora debe retornar “True” si el dato se encuentra repetido en la lista (más de una vez), y “False” en caso contrario

```python
Milista = [3, 6, 9, 12, 9]
numbers = [5, 9]

for i in range(len(numbers)):
  cont = 0
  encontrado = False
  for j in range(len(Milista)):
    if numbers[i] == Milista[j]:
      cont += 1
    if cont >= 2:
      encontrado = True
  print(encontrado)
```

## Ejercicio 6

Realizar un programa que, dada una cantidad, construya una lista de esa cantidad de números enteros, de forma aleatoria y muestre la lista.

```python
from random import randint
n = int(input("Ingrese cuántos números aleatorios quiere que halla en la lista: "))
Milista = []

for i in range(n):
  number = randint(0, 10)
  Milista.append(number)

print(f"La lista de números es {Milista}")
```

## Ejercicio 7

Realizar un programa que, dada una lista de números enteros y un número dado por el usuario, muestre en pantalla los números mayores al digitado por el usuario.

```python
Milista = [3, 6, 9, 12]
ListaNueva = []

n = int(input("Ingresa un número para compararlo con los de la lista: "))

for i in range(len(Milista)):
  if n < Milista[i]:
    ListaNueva.append(Milista[i])
    print(f"El número {Milista[i]} es mayor al número {n}")
print(f"La lista de números es la siguiente {ListaNueva}")
```

## Ejercicio 8

Realizar un programa, que recibe una lista de números enteros, genere como resultado dos listas, una con los elementos pares y otra con los elementos impares.

```python
Milista = [3, 6, 9, 12]
ListaPar = []
ListaImpar = []

for i in Milista:
  if i % 2 == 0:
    ListaPar.append(i)
  else:
    ListaImpar.append(i)
print(f"Lista Par: {ListaPar}")
print(f"Lista Impar: {ListaImpar}")
```

## Ejercicio 9

Realizar un programa que dadas 2 listas, una con 10 nombres y otra con 10 teléfonos, como resultado debe generar una única lista que posee combinados cada nombre con teléfono.

```python
nombres = ["Arnold", "Miguel", "Andrés", "Sebastián"]
teléfonos = [3023844918, 3122941075, 3116193387, 3157982243]
lista_combinada = []

for i in range(len(nombres)):
  variable = nombres[i] + " - " + str(teléfonos[i])
  lista_combinada.append(variable)

print(f"La lista combinada es {lista_combinada}")
```
