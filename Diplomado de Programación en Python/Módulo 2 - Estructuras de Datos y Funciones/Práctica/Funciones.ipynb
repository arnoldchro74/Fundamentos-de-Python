## Ejercicio 1

Realizar un menú donde se le pide al usuario 2 valores. El programa por cada opción del menú debe usar una función o procedimiento independiente para cada una de las siguientes operaciones. (El menú debe tener una única opción para salir de él y se debe repetir hasta que el usuario digite la opción de salida):

- Suma de los valores.
- Resta de los valores.
- Multiplicación de los valores.
- División de los valores (validar que no se incurra en una división por cero).
- Potencia de los valores (el usuario debe elegir cuál de los 2 valores es la base y cuál es la potencia).
- Verificar cuál valor es el mayor o si son iguales

```python
# Definimos las funciones

def suma(a, b):
  resultado = a + b
  return resultado

def resta(a, b):
  resultado = a - b
  return resultado

def multiplicación(a, b):
  resultado = a * b
  return resultado

def división(a, b):
  if b != 0:
    resultado = a / b
  return resultado

def potencia(a, b):
  resultado = a ** b
  return resultado

def mayor(a, b):
  if a > b:
    resultado = a
  elif a == b:
    resultado = "son iguales"
  else:
    resultado = b
  return resultado

condición = True

while condición == True:
  print("-----Menú-----")
  print("Este es un Menú de opciones en las que escogerás \nque operación matemática utilizar")
  print("Si quieres salir del Menú, escribe No")
  respuesta = input("¿Quieres continuar con el Menú? (Sí/No): ").strip().lower()
  if respuesta == "no":
    condición = False
  else:
    print("Escoge que opción del Menú quieres indicando el número de la opción")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Potencia")
    print("6. Mayor")
    número_menu = int(input("Digita el número de la opción del menú: "))
    a = float(input("Escoge el número mayor: "))
    b = float(input("Escoge el número menor: "))

    if número_menu == 1:
      resultado = suma(a, b)
      print(f"El resultado de la suma es {resultado}")
    elif número_menu == 2:
      resultado = resta(a, b)
      print(f"El resultado de la resta es {resultado}")
    elif número_menu == 3:
      resultado = multiplicación(a, b)
      print(f"El resultado de la multiplicación es {resultado}")
    elif número_menu == 4:
      resultado = división(a, b)
      print(f"El resultado de la división es {resultado}")
    elif número_menu == 5:
      número_base = input("Escribe si la base es a o b: ").lower()
      if número_base == "a":
        resultado = potencia(a, b)
      else:
        resultado = potencia(b, a)
      print(f"El resultado de la potencia es {resultado}")
    elif número_menu == 6:
      resultado = mayor(a, b)
      print(f"El número mayor es {resultado}")
```

## Ejercicio 2

¡Ahora tú! Realiza un programa que tendrá 3 funciones:
- Una de ella es la función main() donde le pediremos al usuario los datos necesarios para calcular el área de un rectángulo (largo y ancho). Recordemos que el área es largo*ancho.  

- En una segunda función que llamaremos calcularAreaRectangulo(….) debemos recibir los datos desde el main y con ellos hacer el cálculo del área y retornarlo como resultado de la función.

- Finalmente la tercer función se llamará impreRectángulo(....) y ella se encarga de llamar a la función calcularAreaRectangulo(….), ese dato lo imprime en la pantalla con un print().

No olvides invocar la función main() para que todo pueda empezar

```python
def main():
  w = float(input("Escribe el ancho del rectángulo: "))
  l = float(input("Escribe el largo del rectángulo: "))
  return w, l

def calcularAreaRectangulo():
  w, l = main()
  area = w * l
  return area

def impreRectángulo():
  area = calcularAreaRectangulo()
  print(f"El área del rectángulo es {area}")

impreRectángulo()
```
