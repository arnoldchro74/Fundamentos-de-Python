# Diccionarios

## Ejercicio 1

Realizar un programa que, dadas 5 notas de cursos digitados por el usuario, se guarden en un diccionario:

```python
dicc = {}

for i in range(5):
  materia = input("Ingresa la materia de la cual obtuviste tu nota: ")
  nota = float(input("Ingresa la nota de la materia: "))
  dicc[materia] = nota

print(f"El diccionario con notas y materia es {dicc}")
```

## Ejercicio 2

Realizar un programa que toma una lista con cédulas y otra lista que contiene nombres de un grupo de usuarios, con esta información se debe construir un diccionario donde la llave es la cédula y el valor es el nombre

```python
cedula = [1007602843, 59125080, 7709269]
nombres = ['Arnold', 'Ana', 'William']

registro = {}

for i in range(len(cedula)):
  registro[cedula[i]] = nombres[i]

print(f"El diccionario de registros es {registro}")
```

## Ejercicio 3

Tomando el ejercicio anterior, transforma el programa para que ahora reciba también una lista con los salarios, ahora el diccionario debe quedar con la llave como la cédula y el valor asociado será una lista que tiene en su primera posición el nombre y en la segunda el salario de cada usuario

```python
cedula = [1007602843, 59125080, 7709269]
nombres = ['Arnold', 'Ana', 'William']
salario = [1850000, 3000000, 12000000]

registro = {}

for i in range(len(cedula)):
  registro[cedula[i]] = [nombres[i], salario[i]]

print(f"El diccionario de registros es {registro}")
```
