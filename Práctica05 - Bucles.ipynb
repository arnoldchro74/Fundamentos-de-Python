## Ejercicio 1

Haz que el usuario introduzca números enteros por teclado. Mientras el usuario no introduzca el 0, muestra si el número introducido es par o impar.

```python
print("Este programa te mostrará si el número introducido por ti es par o impar. Para parar el bucle, debes introducir el 0.")

number = int(input("Introduce un número entero: "))

while number != 0:
  if number % 2 == 0:
    print("El número {} es par.".format(number))
  elif number % 2 == 1:
    print("El número {} es impar.".format(number))
  number = int(input("Introduce otro número entero: "))
else:
  print("Has introducido el número 0. El bucle ha terminado.")
```

## Ejercicio 16

Calcular el factorial de un número. La función factorial $ f (n) = n! $ se define como
el producto de los n primeros enteros positivos; es decir,

$$ f(n) = n! = 1\cdot 2\cdot 3 \cdots (n - 1) \cdot n   $$

**Nota:** Recuerde que una propiedad importante del factorial de un número es,

$$ n! = n (n - 1)! $$



```python
#Este problema tiene dos maneras de resolverse, con un for o un while
print("Este algoritmo te arrojara el factorial del número que indiques.")

n = int(input("Ingresa el número a calcularle el factorial: "))
prod = 1
```

```python
# Bucle for
for i in range(1, n + 1):
  prod *= i
else:
  print("El resultado del factorial del número {} es {}.".format(n, prod))
```

```python
# Bucle while
count = 1

while count <= n:
  prod *= count
  count += 1
else:
  print("El resultado del factorial del número {} es {}.".format(n, prod))
```

## Ejercicio 25

Crea un programa que invierta el orden de las cifras de un número a través de divisiones sucesivas.

*PISTA:* Si se halla el residuo de dividir un número entre 10 tenemos la cifra de las unidades…


```python
print("Este algoritmo invertirá el orden de las cifras del número que suministres.")

num = float(input("Ingresa el número a invertir las cifras: "))
figures = num
num_screen = num
count1 = 0
count2 = 1
new_num = 0

while figures != 0:
  figures //= 10
  count1 += 1
else:
  while count2 <= count1:
    companion = float("1" + ((count1 - count2) * "0"))
    new_num += num % 10 * companion
    num //= 10
    count2 += 1
  print("El número de digitos del número es {}.".format(count1))
  print("El número invertido de digitos del número {} es {}.".format(num_screen, new_num))
```

Se debe mostrar en pantalla cuales son las concentraciones obtenidas cada vez que se prepara la solución hasta que se cumpla la condición, se debe mostrar al final cuantas veces se debió repetir la solución.

```python
from random import randint

etanol = 16
attempts = 0

while etanol > 15:
  water = randint(0, 100)
  rest = 100 - water
  etanol = rest / 2
  acetic_acid = rest / 2
  attempts += 1

  if water < 100:
    if etanol <= 15:
      print("Las concentraciones obtenidas son las siguientes: agua {}%, etanol {}%, ácido acético {}%.".format(water, etanol, acetic_acid))
      print("Se tuvo {} intento(s).".format(attempts))
      break
    else:
     print("Las concentraciones obtenidas de etanol son muy altas y eso es peligroso para su seguridad. Se debe repetir el experimento. \nLas concentraciones son de: agua {}%, etanol {}%, ácido acético {}%.".format(water, etanol, acetic_acid))
  else:
   print("La cantidad de agua suministrada es superior al frasco a emplear y, por tanto, no se puede suministrar etanol ni ácido acético.")
```

## Ejercicio 28

Una compañía paga semanalmente $ \$500000$ a sus vendedores más una comisión de 9% del total de ventas realizadas. Elabore el pseudocódigo y diagrama de flujo que pida al usuario el total de ventas y determine el pago que deberá realizarse a cada vendedor. Mostrar cuánto fue el pago y a cuántos trabajadores les pagaron. Al finalizar el gerente desea saber a cuantos vendedores se les pagó más de $\$800000$.

```python
print("Este algoritmo tiene la finalidad de calcular cuánto se le debe pagar a cada empleado de acuerdo al total de ventas realizadas \npor este mismo.")

num_sellers = int(input("Indique a cuántos trabajadores se les va a pagar: "))

count = 1
count1 = 0
count2 = 0

while count <= num_sellers:
  print("Trabajador {}".format(count))
  sales_for_sellers = float(input("Indique cuántas ventas realizo el trabajador correspondiente: "))
  payment_to_sellers = 500000 + 0.09 * sales_for_sellers
  if payment_to_sellers > 800000:
    count1 += 1
  else:
    count2 += 1
  count += 1
else:
  print("Finalmente, se le pagó más de $800.000 a {} empleados y a {} se les pagó $800.000 o menos.".format(count1, count2))
```

## Ejercicio 29

El curso de Fotografía se califica con dos parciales (el primero tiene un peso de 30% y el segundo 35%), una nota de laboratorios (25%) y una nota del trabajo final del curso (10%). Calcular la nota definitiva de cada uno de los estudiantes para un grupo de n estudiantes.

```python
print("Este algoritmo tiene la finalidad de calcular la nota del curso de Fotografía para el número de estudiantes que usted indique.")

n = int(input("Indique el número de estudiantes a calificar: "))

for i in range(n):
  print("Estudiante {}".format(i+1))
  first_partial = float(input("Ingrese la nota del parcial 1: "))
  second_partial = float(input("Ingrese la nota del parcial 2: "))
  lab = float(input("Ingrese la nota de los laboratorios: "))
  final_work = float(input("Ingrese la nota del trabajo final: "))
  course_note = first_partial * 0.3 + second_partial * 0.35 + lab * 0.25 + final_work * 0.1
  print("La nota del estudiante {} es {}.".format(i+1, course_note))
```

Recuerde que una progresión aritmética (PA) es definida como una sucesión en la que la diferencia entre cualquier término y el anterior es la misma a lo largo de toda la sucesión. La diferencia algebraica entre cada término y el anterior viene denotada por $d$ y el primer término se expresa como $a$ para $n$ términos.

El n-ésimo término viene denotado por la fórmula general,

$$ T_{n} = a + (n-1)d$$

Y el n-ésimo término viene denotado por la fórmula recursiva,

$$ T_{n} = T_{n-1} + d$$

para $n \geq  2 $ y $ T_{1} = a $

```python
a = float(input("Ingresa el término inicial a: "))
d = float(input("Ingresa la diferencia d: "))
n = int(input("Ingresa el término deseado de la sucesión: "))
cont = 1
T = 0

while cont <= n:
  T = a + (cont - 1) * d
  print("El término {} es {}.".format(cont, T))
  cont += 1
```

## Ejercicio 38

Realice un algoritmo que se encargue de hacer la suma de $n$ términos de una progresión geométrica. Recuerde que la suma de $n$ términos de una PG finita viene dada por la expresión,


$$ S_{n} = \frac{a (1 - r^{n})}{1-r}  $$


Y, cuando se posee una sucesión infinita,

$$ S_{n} = \frac{a}{1-r}  $$

con tal de que $ -1 < r < 1 $.


**Nota:** Recuerde que dependiendo de si es una sucesión finita o infinita la fórmula puede cambiar.

```python
a = float(input("Ingresa el término inicial a: "))
r = float(input("Ingresa la razón r: "))
answer = input("Ingresa si se tiene una progresión geométrica finita o infinita: ")
answer = answer.lower()

if answer == "infinita":
  S = a / (1 - r)
  print("Se tiene que la suma de la progresión geométrica infinita es {}.".format(S))
elif answer == "finita":
  n = int(input("Ingresa el término de la sucesión hasta el cual se hará la suma: "))
  cont = 1
  T = 0
  S = 0
  while cont <= n:
    T = a * r**(cont - 1)
    S += T
    print("El término {} es {}.".format(cont, T))
    cont += 1
  print("La suma de la progresión geométrica finita es {}.".format(S))
```
