# Tuplas en Python

## Ejercicio 1

Pide al usuario el número de números enteros que va a introducir por teclado. Para cada uno de esos
números, crea una tupla donde la primera entrada sea el número entero y, la segunda, la palabra “par” o
“impar” según la paridad del número entero. Muestra la tupla recién creada al usuario.

```python
print("Este algoritmo te mostrará si el número que ingreses es par o impar.")

n = int(input("Ingresa la cantidad de números que vas a digitar: "))
numbers = []

for i in range(n):
  num = int(input("Ingresa el número correspondiente: "))
  if num % 2 == 0:
    tupla = (num, "par")
    print(tupla)
    numbers.append(tupla)
  else:
    tupla = (num, "impar")
    print(tupla)
    numbers.append(tupla)

print("La lista de tuplas ordenadas es", numbers)
```

## Ejercicio 2

Dado un año proporcionado por el usuario, crea una tupla de dos elementos cuya primera entrada sea el año
y, la segunda entrada, el horóscopo chino correspondiente.

```python
print("Este algoritmo te indicará cuál es el animal que corresponde con el año de acuerdo con el horóscopo chino.")

horos_china = ["Mono", "Gallo", "Perro", "Cerdo", "Rata", "Buey", "Conejo", "Dragón", "Serpiente", "Caballo", "Cabra"]

year = int(input("Introduce el año: "))

ind = year % 12

tupla = (year, horos_china[ind])

print("El animal correspondiente al año es", tupla)
```

## Ejercicio 3

Dada una frase proporcionada por el usuario, crea una lista de tuplas indicando palabra, longitud de cada
palabra, letra inicial y posición que ocupan dentro de la frase.

```python
print("Este algoritmo te pedirá una frase y te entregará un resumen de cada palabra en una lista con tuplas.")

phrase = input("Introduce la frase: ")
cont = 1
l = []

for i in phrase.split():
  t = (i, len(i), i[0], cont)
  l.append(t)
  cont += 1

print("El resumen de la frase palabra por palabra es", l)
```

## Ejercicio 4

Haz que el usuario introduzca palabras hasta que introduzca una palabra vacía. Guarda todas las palabras
en una tupla y muestra la primera y la última introducidas haciendo uso del método unpacking.

**PISTA:** Para guardar los elementos de uno en uno vas a tener que utilizar un tipo de dato que no es tupla
y luego transformarlo a tupla.

```python
print("Este algoritmo te entregará la primera y la última palabra de las palabras que introduzcas.")
print("Para parar el bucle, sólo oprime enter para tener un string vacío. Para continuar, escribe 'sí'")

ans = "sí"
l = []

while ans != "":
  word = input("Introduce una palabra: ")
  l.append(word)
  ans = input("¿Quieres introducir otra palabra (sí es 'sí', ' ' es no): ")
  ans.lower()

words = tuple(l)

word1, *_, word2 = words

print("La primera palabra fue {} y la última palabra fue {}.".format(word1, word2))
```

## Ejercicio 5

Dada una lista de palabras, crea otra lista del mismo tamaño que guarde la longitud de cada palabra. Usa
la función zip() para crear un diccionario con claves las palabras y valores, su longitud.

```python
print("Este algoritmo te indicará la longitud de cada palabra que ingreses.")

print("Para parar el bucle, sólo escribe 'no'. Para continuar, escribe 'sí'")

ans = "sí"
l = []
q = []

while ans != "no":
  word = input("Introduce una palabra: ")
  l.append(word)
  ans = input("¿Quieres introducir otra palabra (sí es 'sí', ' ' es no): ")
  ans.lower

words = tuple(l)

for i in l:
  q.append(len(i))

length = tuple(q)

dicc = dict(zip(words, length))

print("El diccionario con las palabras y su longitud es", dicc)
```

## Ejercicio 6

Haz que el usuario introduzca palabras hasta que introduzca una palabra vacía. Guarda todas las palabras
en una tupla y muestra la tupla y el número total de caracteres que ha introducido.

**PISTA:** Para guardar los elementos de uno en uno vas a tener que utilizar un tipo de dato que no es tupla
y luego transformarlo a tupla.

```python
print("Este algoritmo te entregará cada palabra que introduzcas y contará al final el total de caracteres de cada palabra.")
print("Para parar el bucle, sólo oprime enter para tener un string vacío. Para continuar, escribe 'sí'")

ans = "sí"
l = []
caract = 0

while ans != "":
  word = input("Introduce una palabra: ")
  l.append(word)
  ans = input("¿Quieres introducir otra palabra (sí es 'sí', ' ' es no): ")
  ans.lower()

for i in l:
  caract += len(i)

words = tuple(l)

print("El conjunto de palabras introducido este", words, "y su cantidad de caracteres es", caract)
```

## Ejercicio 7

Crea una lista de 20 tuplas de tamaño 2. La primera entrada será un número entero entre 1 y 20 y la segunda
entrada contendrá una lista con los 10 primeros múltiplos del número entero correspondiente. Por último,
muestra las tablas de multiplicar del 1 al 20 con el formato “1 x 1 = 1”.

```python
print("Este algoritmo te mostrará las tablas de multiplicar del 1 al 20 haciéndo uso de tuplas.")

numbers = []
multiples = []

for i in range(20):
  numbers.append(i + 1)
  multiples.append([])
  for j in range(10):
    multiples[i].append((i + 1) * (j + 1))

for i in range(20):
  for j in range(10):
      print(i + 1, "x", j + 1, "=", multiples[i][j])

total = list(zip(numbers, multiples))
print("La lista completa con números y sus múltiplos es", total)
```

## Ejercicio 8

Pide al usuario dos números enteros por teclado. Asegúrate de que el primero es mayor o igual al segundo.
Realiza la división entera y guarda en una tupla el dividendo, el divisor, el cociente y el resto de la división
entera realizada y muéstrale al usuario el resultado por pantalla.

```python
print("Este algoritmo guardará el dividento, el divisor, el cociente y el residuo de una división de \ndos números enteros que proporciones.")

a = int(input("Ingresa el número mayor: "))
b = int(input("Ingresa el número menor: "))

if a >= b:
  c = a // b
  r = a % b
  t = (a, b, c, r)
  print("El dividendo es {}; el divisor es {}; el cociente es {} y el residuo es {}.".format(a, b, c, r))
  print("Ordenándolo en forma de tupla es", t)
else:
  print("El número a no es mayor o igual que el número b.")
```

## Ejercicio 9

Pide al usuario números entre 0 y 360. Para cada número, crea una tupla donde la primera entrada sea
dicho número y, la segunda, la medida angular correspondiente en radianes. Recuerda, $360^{\circ} = 2\pi $. En
este caso, utiliza $\pi $ = 3.141592653589793

```python
from numpy import pi

print("Este algortimo se encargará de convertir los números que ingreses en grados y los pasará a radianes. ")

anglesg = []
anglesr = []

ans = "sí"

while ans != "no":
  angle = float(input("Ingresa el ángulo en grados: "))
  if angle >= 0 and angle <= 360:
    rad = (angle * 2 * pi) / 360
    anglesg.append(angle)
    anglesr.append(rad)
  else:
    print("No has introducido un ángulo entre 0 y 360 grados.")
  ans = input("¿Deseas continuar agregando ángulos?: ")
  ans = ans.lower()

angles = list(zip(anglesg, anglesr))

print("La lista con los ángulos en grados y radiantes es {}.".format(angles))
```
