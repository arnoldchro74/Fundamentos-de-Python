## Ejercicio 1

Dado un número entero introducido por teclado, guarda sus divisores en un conjunto y muéstralo.

```python
print("Este algoritmo almacenará todos los divisores del número que suministres en un conjunto y te los mostrará al final.")

myset = set()

n = int(input("Introduce un número entero: "))

for i in range(1, n + 1):
  if n % i == 0:
    myset.add(i)

print("El conjunto de números divisores del número {} suministrado es: ".format(n))
myset
```

## Ejercicio 2

Crea un programa que dado un conjunto, nos devuelva su mínimo. Debes hacerlo sin recurrir a la función
min().

```python
print("Este algoritmo te entregará el valor mínimo del conjunto de números que suministres.")

A = set()
min = None

answer = "sí"

while answer != "no":
  n = float(input("Introduce un número que deseas agregar dentro del conjunto A: "))
  A.add(n)
  answer = input("¿Deseas agregar otro número?: ")
  answer = answer.lower()

for i in A:
  if min is None or min > i:
    min = i

print("El número menor del conjunto {} es {}.".format(A, min))
```

## Ejercicio 3

Dada una frase introducida por teclado, guarda en un conjunto todas las palabras que empiecen por la letra
indicada por el usuario.

```python
print("Este algoritmo guardará las palabras de la frase que suministres que comiencen por la letra que indiques.")

s = input("Introduce la frase: ")
s = s.lower()
l = input("Introduce la letra: ")

myset = set()

for i in s.split():
  if i[0] == l:
    myset.add(i)

print("El conjunto con las palabras de la frase que coinciden con la primera letra entregada es", myset)
```

## Ejercicio 4

Dado un conjunto, crea un programa que nos devuelva el caracter con mayor valor ASCII. Debes hacerlo sin
recurrir a la función max().

```python
#Función que pasa de carácter a número - ord()
print("Este algoritmo te informará cuál número asociado a cada carácter de un string es el mayor.")

answer = "sí"
myset = set()
max = None

while answer != "no":
  c = input("Ingresa un carácter que quieras ingresar al conjunto: ")
  n = ord(c)
  myset.add(n)
  answer = input("¿Deseas ingresar otro carácter? (sí o no): ")
  answer = answer.lower()

for i in myset:
  if max is None or max < i:
    max = i

for i in myset:
  if i == max:
    caract = chr(i)

print("El carácter {} tiene el número más grande, el cual es el {}.".format(caract, max))
```

## Ejercicio 6

indicada por el usuario.

## Ejercicio 7

Dada una frase introducida por teclado, guarda en un conjunto la primera letra de cada palabra sin hacer
uso del método .split().

```python
print("Este algoritmo guardará en un conjunto la primera letra de cada palabra escrita en la frase suministrada por el usuario.")

phrase = input("Ingresa la frase: ")

myset = set()

num = phrase.count(" ")
n = 0

if phrase[0] != " ":
  myset.add(phrase[0])

for i in range(num):
  n = phrase.find(" ", n, -1)
  myset.add(phrase[n + 1])
  n += 1

print("El conjunto de la primera letra de cada palabra es", myset)
```

## Ejercicio 8

Dada una frase introducida por teclado, guarda en un conjunto todas las palabras con longitud par.

```python
print("Este algoritmo guardará en un conjunto todas las palabras que tengan una longitud par.")

myset = set()

phrase = input("Ingresa la frase: ")

for i in phrase.split():
  if len(i) % 2 == 0:
    myset.add(i)

print("El conjunto de palabras con longitud par es", myset)
```

## Ejercicio 9

Dada una frase introducida por teclado, guarda en un conjunto todas las palabras que acaben por la letra
indicada por el usuario.

```python
print("Este algoritmo guardará las palabras de la frase que suministres que finalicen por la letra que indiques.")

s = input("Introduce la frase: ")
s = s.lower()
l = input("Introduce la letra: ")

myset = set()

for i in s.split():
  if i[-1] == l:
    myset.add(i)

print("El conjunto con las palabras de la frase que coinciden con la última letra entregada es", myset)
```

## Ejercicio 10

Dada una frase introducida por teclado, guarda en un conjunto todas las palabras palíndromas.

```python
print("Este algoritmo guardará en un conjunto todas las palabras palíndromas dentro de la frase que suministres.")

myset = set()

s = input("Ingresa la frase: ")
s = s.lower()

for i in s.split():
  word = ""
  for j in range(len(i)):
    c = i[len(i) - j - 1]
    word = word + c
  if i == word:
    myset.add(i)

print("El conjunto de palabras palíndromas dentro de la frase es", myset)
```

## Ejercicio 11

Vamos a crear un nuevo conjunto a partir del conjunto A = {1, 2, 3, 4, 5} cuyos valores sean el cuadrado de los elementos del conjunto A.

```python
print("Este algoritmo elevará al cuadrado los elementos del conjunto A en un nuevo conjunto B.")

A = {1, 2, 3, 4, 5}

B = set()

for i in A:
  B.add(i**2)

print("El nuevo conjunto B con los valores al cuadrado es este", B)
```

## Ejercicio 12

Teniendo en cuenta el conjunto S y el conjunto T, sin usar el operador de intersección &, calcularemos la intersección de dichos conjuntos.

```python
print("Este algoritmo calculará la intersección de los conjuntos S y T sin usar la función .intersection()")
S = set()
T = set()
I = set()

answer1 = "sí"
cont1 = 0
answer2 = "sí"
cont2 = 0

while answer1 != "no":
  element1 = int(input("Introduce el elemento {} del conjunto S: ".format(cont1 + 1)))
  S.add(element1)
  answer1 = input("¿Quieres introducir un nuevo elemento al conjunto S? (sí o no): ")
  answer1 = answer1.lower()
  cont1 += 1

while answer2 != "no":
  element2 = int(input("Introduce el elemento {} del conjunto T: ".format(cont2 + 1)))
  T.add(element2)
  answer2 = input("¿Quieres introducir un nuevo elemento al conjunto T? (sí o no): ")
  answer2 = answer2.lower()
  cont2 += 1

for i in S:
  for j in T:
    if i == j:
      I.add(i)

print("La intersección del conjunto S y el conjunto T es", I)
print("Intersección", S.intersection(T))
```

## Ejercicio 14

Realice un programa que obtenga la unión de dos conjuntos que introduzca el usuario.

```python
print("Este algoritmo te entregará la unión de dos conjuntos A y B (estos conjuntos tendrán números).")

n1 = int(input("Indica la cantidad de elementos que tendrá el conjunto A: "))

A = set()

for i in range(n1):
  number = float(input("Ingresa el elemento {} del conjunto A: ".format(i + 1)))
  A.add(number)

n2 = int(input("Indica la cantidad de elementos que tendrá el conjunto B: "))

B = set()

for i in range(n2):
  number = float(input("Ingresa el elemento {} del conjunto B: ".format(i + 1)))
  B.add(number)

print("La unión entre estos dos conjuntos es", A.union(B))
```

## Ejercicio 15

Realice un programa que elimine un elemento determinado por el usuario de un conjunto.

```python
print("Este algoritmo eliminará un elemento del conjunto A que tú indiques.")

n = int(input("Indica la cantidad de elementos que tendrá el conjunto A: "))

A = set()

for i in range(n):
  number = float(input("Ingresa el elemento {} del conjunto A: ".format(i + 1)))
  A.add(number)

delete = float(input("Indica el elemento que quieres eliminar: "))
A.discard(delete)
print("El conjunto actualizado es", A)
```
