# Functiones Lambda en Python

## Ejercicio 1

Crea una función lambda que dado un número entero multiplique por su anterior y su siguiente. Por ejemplo,
si proporcionamos n = 3, nos tendrá que devolver 2 · 3 · 4 = 24.

```python
print("Este algoritmo multiplicará el número que indiques por su anterior y por el siguiente.")

n = int(input("Ingresa el número: "))

multiplies = lambda x: (x - 1) * x * (x + 1)

print("El producto es {}.".format(multiplies(3)))
```

## Ejercicio 2

Crea una función lambda que dados dos números devuelva si el primero es mayor.

```python
print("Este algoritmo te indicará cuál de los dos números que suministres es mayor.")

a = int(input("Ingresa el número a: "))
b = int(input("Ingresa el número b: "))

compare = lambda x, y: x if x >= y else y
print("El número mayor es {}.".format(compare(a, b)))
```

## Ejercicio 3

Dada una lista de palabras, quédate con filter() con las que tengan más vocales que consonantes. Necesitarás
una función que devuelva si una palabra tiene más vocales que consonantes.

```python
def contv_c(word):
  v = 0
  c = 0

  for i in word.lower():
    if i == "a" or i == "e" or i == "i" or i == "o" or i == "u":
      c += 1
    else:
      v += 1

  if c > v:
    return word

words = ["ayuda", "amigo", "auxilio", "beber", "largo", "melardo"]
result = list(filter(contv_c, words))
result
```

## Ejercicio 4

Dada una lista de números enteros, quédate con filter() con los que tengan más de 5 divisores. Necesitarás
una función que devuelva el número de divisores de un número dado.

```python
def divisors(x):

  div = 0

  for i in range(x + 1):
    if x % (i + 1) == 0:
      div += 1

  return div

numbers = [3, 4, 10, 20, 15, 30]
result = list(filter(lambda x: x if divisors(x) > 5 else None, numbers))
result
```

## Ejercicio 5

Dada una lista de palabras, quédate con reduce() con la palabra más larga. Necesitarás una función que
compare dos palabras y devuelva la que tenga mayor longitud.

```python
from functools import reduce

def may(word1, word2):

  if len(word1) >= len(word2):
    return word1
  else:
    return word2

words = ["Enseñar", "Disfrutar", "Aprender", "Vivir", "Tocar", "Investigar", "Estudiar"]
reduce(may, words)
```

## Ejercicio 6

Dada una lista de palabras, calcula el número de vocales de cada una con map().

```python
def vowels(word):
  contv = 0
  for i in word.lower():
    if i == "a" or i == "e" or i == "i" or i == "o" or i == "u":
      contv += 1
  return contv

words = ["Enseñar", "Disfrutar", "Aprender", "Vivir", "Tocar", "Investigar", "Estudiar"]
list(map(vowels, words))
```

## Ejercicio 7

Dada una lista de palabras, quédate con reduce() y map() con la palabra con más consonantes. Necesitarás
una función que cuente el número de consonantes de una palabra y otra que dados dos números, devuelva el
mayor.

```python
from functools import reduce
def consonants(word):
  contc = 0
  for i in word.lower():
    if i != "a" and i != "e" and i != "i" and i != "o" and i != "u":
      contc += 1
  return contc

def may(a, b):
  if a >= b:
    return a
  return b

words = ["Enseñar", "Disfruta", "Aprender", "Vivir", "Tocar", "Investigar", "Estudiar"]
numbers_count = list(map(consonants, words))
result = reduce(may, numbers_count)
for i in range(len(words)):
  if consonants(words[i]) == result:
    special_word = words[i]
print("La palabra con más consonantes es", special_word)
```

## Ejercicio 9

Dada una lista de números reales, ordénalos con sorted() por valor absoluto de menor a mayor.

```python
numbers = [-4, 2, -3, 8, 5, 6, 20, 15, -25]

result = sorted(numbers, key = lambda x: x if x > 0 else -1 * x, reverse = False)
final_result = list(map(lambda x: x if x > 0 else -1 * x, result))
print(final_result)
print(result)
```
