*  La propiedad .module que devuelva el módulo del vector. Recuerda que el módulo de un vector 2D se
calcula con raices. Para calcular raíces cuadradas, dispones del método math.sqrt(). Tiene que quedar $ \sqrt{x^{2} + y^{2}}$.

```python
import math as mt

class Vector2D:
	def __init__(self, x, y):
		self.x = x
		self.y = y

	@property
	def module(self):
		return mt.sqrt(self.x**2 + self.y**2)

	def scalar_prod(self, scalar = 1):
		vector = [self.x, self.y]
		values = []
		for i in vector:
			values.append(i * scalar)
		return values

	def __str__(self):
		return "({}, {})".format(self.x, self.y)
```

## Ejercicio 3

Crea los siguientes métodos:

*   El método de clase .sum() que dados dos vectores los sume y devuelva un objeto de la clase Vector2D.
*   El método de clase .subtract() que dados dos vectores los reste y devuelva un objeto de la clase
Vector2D.

```python
import math as mt

class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    @property
    def module(self):
        return mt.sqrt(self.x**2 + self.y**2)

    def scalar_prod(self, scalar=1):
        vector = [self.x, self.y]
        values = []
        for i in vector:
            values.append(i * scalar)
        return values

    def __str__(self):
        return "({}, {})".format(self.x, self.y)

    @classmethod
    def sum(cls, vector1: Vector2D, vector2: Vector2D) -> Vector2D:
        x_final = vector1.x + vector2.x
        y_final = vector1.y + vector2.y
        return cls(x_final, y_final)

    @classmethod
    def subtract(cls, vector1: Vector2D, vector2: Vector2D) -> Vector2D:
        x_final = vector1.x - vector2.x
        y_final = vector1.y - vector2.y
        return cls(x_final, y_final)

v1 = Vector2D(1, 2)
v2 = Vector2D(3, 4)
print(Vector2D.sum(v1, v2))
print(Vector2D.subtract(v1, v2))
```

## Ejercicio 4

Crea los siguientes módulos




* El método de clase .distance() que dados dos vectores calcule la distancia entre ellos. Recuerda que
la distancia entre 2 vectores 2D u y v se calcula como $$
\sqrt{(u_{x}− v_{x})^{2} + (u_{y} − v_{y})^{2}}$$
*  El método estático .dot_product() que dados dos vectores calcule su producto escalar. Recuerda que
el producto escalar de 2 vectores 2D u y v se calcula como $$u · v = u_{x}v_{x} + u_{y}v_{y}$$



```python
import math as mt

class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    @property
    def module(self):
        return mt.sqrt(self.x**2 + self.y**2)

    def scalar_prod(self, scalar=1):
        vector = [self.x, self.y]
        values = []
        for i in vector:
            values.append(i * scalar)
        self.x = values[0]
        self.y = values[1]
        return self.x, self.y

    def __str__(self):
        return "({}, {})".format(self.x, self.y)

    @classmethod
    def sum(cls, vector1: Vector2D, vector2: Vector2D) -> Vector2D:
        x_final = vector1.x + vector2.x
        y_final = vector1.y + vector2.y
        return cls(x_final, y_final)

    @classmethod
    def subtract(cls, vector1: Vector2D, vector2: Vector2D) -> Vector2D:
        x_final = vector1.x - vector2.x
        y_final = vector1.y - vector2.y
        return cls(x_final, y_final)

    @staticmethod
    def dot_product(vector1, vector2):
        value = 0
        for i in range(len(vector1)):
            value += vector1[i] * vector2[i]
        return value

    @classmethod
    def distance(cls, vector1: list, vector2: list) -> Vector2D:
        vector1 = cls(vector1[0], vector1[1])
        vector2 = cls(vector2[0], vector2[1])
        d = mt.sqrt((vector1.x - vector2.x)**2 + (vector1.y - vector2.y)**2)
        return d

v1 = Vector2D(1, 2)
v2 = Vector2D(3, 4)
print(Vector2D.dot_product([1, 2], [3, 4]))
print(Vector2D.distance([1, 2], [3, 4]))
```

*   el método de instancia .scalar_prod()
*   los métodos de clase .sum() y .subtract() para que devuelvan objetos de la clase Vector3D.


*  el método estático .dot_product(), pues ahora el producto escalar se calcula como $$u · v = u_{x}v_{x} + u_{y}v_{y} +  u_{z}v_{z}$$



*   el método de clase .distance(), pues ahora la distancia se calcula como
$$
\sqrt{(u_{x}− v_{x})^{2} + (u_{y} − v_{y})^{2} + (u_{z} − v_{z})^{2}}$$

Recuerda que dispones del método .super() para evitar repeticiones innecesarias de código.

```python
class Vector3D(Vector2D):
    def __init__(self, x, y, z):
        super().__init__(x, y)
        self.z = z

    @property
    def module(self):
        return mt.sqrt(self.x**2 + self.y**2 + self.z**2)

    def scalar_prod(self, scalar=1):
        vector = [self.x, self.y, self.z]
        values = []
        for i in vector:
            values.append(i * scalar)
        self.x = values[0]
        self.y = values[1]
        self.z = values[2]
        return self.x, self.y, self.z


    def __str__(self):
        return "({}, {}, {})".format(self.x, self.y, self.z)

    @classmethod
    def sum(cls, vector1, vector2):
        x_final = vector1[0] + vector2[0]
        y_final = vector1[1] + vector2[1]
        z_final = vector1[2] + vector2[2]
        return cls(x_final, y_final, z_final)

    @classmethod
    def subtract(cls, vector1, vector2):
        x_final = vector1[0] - vector2[0]
        y_final = vector1[1] - vector2[1]
        z_final = vector1[2] - vector2[2]
        return cls(x_final, y_final, z_final)

    @staticmethod
    def dot_product(vector1, vector2):
        value = 0
        for i in range(len(vector1)):
          value += vector1[i] * vector2[i]
        return value

    @classmethod
    def distance(cls, vector1, vector2):
        d = mt.sqrt((vector1[0] - vector2[0])**2 + (vector1[1] - vector2[1])**2 + (vector1[2] - vector2[2])**2)
        return d

v1 = Vector3D(1, 2, 3)
v2 = Vector3D(3, 4, 5)
print(Vector3D.sum([1, 2, 3], [3, 4, 5]))
print(Vector3D.subtract([1, 2, 3], [3, 4, 5]))
print(v1.module)
print(v1.scalar_prod(2))
```

```python
import math as mt

class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    @property
    def module(self):
        return mt.sqrt(self.x**2 + self.y**2)

    def scalar_prod(self, scalar=1):
        vector = [self.x, self.y]
        values = []
        for i in vector:
            values.append(i * scalar)
        self.x = values[0]
        self.y = values[1]
        return self.x, self.y

    def __str__(self):
        return "({}, {})".format(self.x, self.y)

    @classmethod
    def sum(cls, vector1: Vector2D, vector2: Vector2D) -> Vector2D:
        x_final = vector1.x + vector2.x
        y_final = vector1.y + vector2.y
        return cls(x_final, y_final)

    @classmethod
    def subtract(cls, vector1: Vector2D, vector2: Vector2D) -> Vector2D:
        x_final = vector1.x - vector2.x
        y_final = vector1.y - vector2.y
        return cls(x_final, y_final)

    @staticmethod
    def dot_product(vector1, vector2):
        value = 0
        for i in range(len(vector1)):
            value += vector1[i] * vector2[i]
        return value

    @classmethod
    def distance(cls, vector1: list, vector2: list) -> Vector2D:
        vector1 = cls(vector1[0], vector1[1])
        vector2 = cls(vector2[0], vector2[1])
        d = mt.sqrt((vector1.x - vector2.x)**2 + (vector1.y - vector2.y)**2)
        return d

    def extend_to_3D(self, z = 0):
      return Vector3D(self.x, self.y, z)

class Vector3D(Vector2D):
    def __init__(self, x, y, z):
        super().__init__(x, y)
        self.z = z

    @property
    def module(self):
        return mt.sqrt(self.x**2 + self.y**2 + self.z**2)

    def scalar_prod(self, scalar=1):
        vector = [self.x, self.y, self.z]
        values = []
        for i in vector:
            values.append(i * scalar)
        self.x = values[0]
        self.y = values[1]
        self.z = values[2]
        return self.x, self.y, self.z

    def __str__(self):
        return "({}, {}, {})".format(self.x, self.y, self.z)

    @classmethod
    def sum(cls, vector1, vector2):
        x_final = vector1.x + vector2.x
        y_final = vector1.y + vector2.y
        z_final = vector1.z + vector2.z
        return cls(x_final, y_final, z_final)

    @classmethod
    def subtract(cls, vector1, vector2):
      x_final = vector1.x - vector2.x
      y_final = vector1.y - vector2.y
      z_final = vector1.z - vector2.z
      return cls(x_final, y_final, z_final)

    @staticmethod
    def dot_product(vector1, vector2):
        value = 0
        for i in range(len(vector1)):
           value += vector1[i] * vector2[i]
        return value

    @classmethod
    def distance(cls, vector1, vector2):
        d = mt.sqrt((vector1[0] - vector2[0])**2 + (vector1[1] - vector2[1])**2 + (vector1[2] - vector2[2])**2)
        return d

    @classmethod
    def zero(cls):
      l = [0, 0, 0]
      return cls(l[0], l[1], l[2])

    @classmethod
    def horizontal(cls):
      l = [1, 0, 0]
      return cls(l[0], l[1], l[2])

    @classmethod
    def vertical(cls):
      l = [0, 1, 0]
      return cls(l[0], l[1], l[2])

    @classmethod
    def forward(cls):
      l = [1, 0, 1]
      return cls(l[0], l[1], l[2])

rect3 = Vector3D.horizontal()
print(rect3)
```
