```python
def fill_with_zeros(number, digits):
  num = str(number)
  while len(num) <= digits:
    num = '0' + num
  return num

class Date:
  def __init__(self, day = 1, month = 1, year = 1):
    self.day = day
    self.month = month
    self.year = year

  def __str__(self):
    day = fill_with_zeros(self.day, 2 - len(str(self.day)))
    month = fill_with_zeros(self.month, 2 - len(str(self.month)))
    year = fill_with_zeros(self.year, 4 - len(str(self.year)) if self.year < 10 else 5 - len(str(self.year)) if self.year < 100 else 6 - len(str(self.year)))
    return f'{day}/{month}/{year}'

  def isLeap(self):
    if self.year % 4 == 0:
      if self.year % 100 == 0:
        if self.year % 400 == 0:
          print("El año {} es bisiesto".format(self.year))
        else: print("El año {} no es bisiesto".format(self.year))
      else: print("El año {} es bisiesto".format(self.year))
    else: print("El año {} no es bisiesto".format(self.year))

  def totalMonthDays(self):
    if self.month == 2:
      if self.isLeap():
        return 29
      else: return 28
    else:
      if self.month == 4 or self.month == 6 or self.month == 9 or self.month == 11:
        return 30
      else: return 31

  def validDate(self):
    if self.day > self.totalMonthDays():
      print("¡¡¡La fecha introducida no es una fecha válida!!!")
    else:
      if self.month < 1 or self.month > 12:
        print("¡¡¡La fecha introducida no es una fecha válida!!!")
      else:
        if self.day == 0 or self.month == 0 or self.year == 0:
          print("¡¡¡La fecha introducida no es una fecha válida!!!")
        else:
          print("La fecha introducida es válida")

  def monthName(self):
    meses = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    return meses[self.month - 1]

  @staticmethod
  def areEqual(date1, date2):
    if date1.day == date2.day and date1.month == date2.month and date1.year == date2.year:
      return True
    else: return False

  @staticmethod
  def isLater(date1, date2):
    if date1.year > date2.year:
      return True
    elif date1.year == date2.year:
      if date1.month > date2.month:
        return True
      elif date1.month == date2.month:
        if date1.day > date2.day:
          return True
        else: return False
      else: return False
    return False

  @staticmethod
  def isPrevious(date1, date2):
    if date1.year == date2.year:
      if date1.month == date2.month:
        if date1.day == date2.day:
          return f"Las fechas son iguales"
        else:
          if date1.day > date2.day:
            return f"La segunda fecha es previa a la primera"
          else: return f"La primera fecha es previa a la segunda"
      else:
        if date1.month > date2.month:
          return f"La segunda fecha es previa a la primera"
        else: return f"La primera fecha es previa a la segunda"
    else:
      if date1.year > date2.year:
        return f"La segunda fecha es previa a la primera"
      else: return f"La primera fecha es previa a la segunda"

  @classmethod
  def firstDayOfTheYear(cls, year):
    return cls(1, 1, year)

  @classmethod
  def lastDayOfTheYear(cls, year):
    return cls(31, 12, year)

  def plusDay(self):
    if self.day == self.totalMonthDays():
      if self.month == 12:
        self.year += 1
        self.month = 1
        self.day = 1
      else: self.month += 1
      self.day = 1
    else: self.day += 1

  def minusDay(self):
    if self.day == 1:
      if self.month == 1:
        self.year -= 1
        self.month = 12
        self.day = 31
      else: self.month -= 1
      self.day = self.totalMonthDays()
    else: self.day -= 1

  @classmethod
  def copy(cls, date):
    return cls(date.day, date.month, date.year)

  @staticmethod
  def difference(date1, date2):
    if Date.areEqual(date1, date2):
      return 0
    else:
      if Date.isLater(date1, date2):
        count = 0
        d2_copy = Date.copy(date2)
        while not Date.areEqual(date1, d2_copy):
          count += 1
          d2_copy.plusDay()
        return count
      else:
        count = 0
        d2_copy = Date.copy(date2)
        while not Date.areEqual(date1, d2_copy):
          count += 1
          d2_copy.minusDay()
        return count

fecha = Date(4, 11, 2003)
fecha2 = Date(15, 12, 2003)
print(Date.difference(fecha, fecha2))
```

```python
def fill_with_zeros(number, digits):
  num = str(number)
  while len(num) <= digits:
    num = '0' + num
  return num

class Date:
  def __init__(self, day = 1, month = 1, year = 1):
    self.day = day
    self.month = month
    self.year = year

  def __str__(self):
    day = fill_with_zeros(self.day, 2 - len(str(self.day)))
    month = fill_with_zeros(self.month, 2 - len(str(self.month)))
    year = fill_with_zeros(self.year, 4 - len(str(self.year)) if self.year < 10 else 5 - len(str(self.year)) if self.year < 100 else 6 - len(str(self.year)))
    return f'{day}/{month}/{year}'

  def isLeap(self):
    if self.year % 4 == 0:
      if self.year % 100 == 0:
        if self.year % 400 == 0:
          print("El año {} es bisiesto".format(self.year))
        else: print("El año {} no es bisiesto".format(self.year))
      else: print("El año {} es bisiesto".format(self.year))
    else: print("El año {} no es bisiesto".format(self.year))

  def totalMonthDays(self):
    if self.month == 2:
      if self.isLeap():
        return 29
      else: return 28
    else:
      if self.month == 4 or self.month == 6 or self.month == 9 or self.month == 11:
        return 30
      else: return 31

  def validDate(self):
    if self.day > self.totalMonthDays():
      print("¡¡¡La fecha introducida no es una fecha válida!!!")
    else:
      if self.month < 1 or self.month > 12:
        print("¡¡¡La fecha introducida no es una fecha válida!!!")
      else:
        if self.day == 0 or self.month == 0 or self.year == 0:
          print("¡¡¡La fecha introducida no es una fecha válida!!!")
        else:
          print("La fecha introducida es válida")

  def monthName(self):
    meses = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    return meses[self.month - 1]

  @staticmethod
  def areEqual(date1, date2):
    if date1.day == date2.day and date1.month == date2.month and date1.year == date2.year:
      return True
    else: return False

  @staticmethod
  def isLater(date1, date2):
    if date1.year > date2.year:
      return True
    elif date1.year == date2.year:
      if date1.month > date2.month:
        return True
      elif date1.month == date2.month:
        if date1.day > date2.day:
          return True
        else: return False
      else: return False
    return False

  @staticmethod
  def isPrevious(date1, date2):
    if date1.year == date2.year:
      if date1.month == date2.month:
        if date1.day == date2.day:
          return f"Las fechas son iguales"
        else:
          if date1.day > date2.day:
            return f"La segunda fecha es previa a la primera"
          else: return f"La primera fecha es previa a la segunda"
      else:
        if date1.month > date2.month:
          return f"La segunda fecha es previa a la primera"
        else: return f"La primera fecha es previa a la segunda"
    else:
      if date1.year > date2.year:
        return f"La segunda fecha es previa a la primera"
      else: return f"La primera fecha es previa a la segunda"

  @classmethod
  def firstDayOfTheYear(cls, year):
    return cls(1, 1, year)

  @classmethod
  def lastDayOfTheYear(cls, year):
    return cls(31, 12, year)

  def plusDay(self):
    if self.day == self.totalMonthDays():
      if self.month == 12:
        self.year += 1
        self.month = 1
        self.day = 1
      else: self.month += 1
      self.day = 1
    else: self.day += 1

  def minusDay(self):
    if self.day == 1:
      if self.month == 1:
        self.year -= 1
        self.month = 12
        self.day = 31
      else: self.month -= 1
      self.day = self.totalMonthDays()
    else: self.day -= 1

  @classmethod
  def copy(cls, date):
    return cls(date.day, date.month, date.year)

  @staticmethod
  def difference(date1, date2):
    if Date.areEqual(date1, date2):
      return 0
    else:
      if Date.isLater(date1, date2):
        count = 0
        d2_copy = Date.copy(date2)
        while not Date.areEqual(date1, d2_copy):
          count += 1
          d2_copy.plusDay()
        return count
      else:
        count = 0
        d2_copy = Date.copy(date2)
        while not Date.areEqual(date1, d2_copy):
          count += 1
          d2_copy.minusDay()
        return count

  @classmethod
  def randomDate(cls):
    import random
    month = random.randint(1, 12)
    year = random.randint(0, 3000)
    random_date = cls(1, month, year)
    random_date.day = random.randint(1, random_date.totalMonthDays())
    return random_date

fecha = Date.randomDate()
print(fecha)
```
